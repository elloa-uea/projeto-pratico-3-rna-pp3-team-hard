### 4.1 Baseando-se nos fundamentos de visualização de dados abordados na disciplina, plote cada um dos atributos preditores de maneira a enfatizar a sua distribuição, tendência central e dispersão

- A seguir tem-se a plotagem de três gráficos (Histograma, Scatter e Boxplot) para cada atributo quantitativo da base de dados.

```python
for col_name in base.columns:
    if 'Soil' not in col_name \
        and 'Wildern' not in col_name \
            and 'Cover_' not in col_name:
        #col_name = 'Elevation'
        fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(20,5))
        fig.suptitle('Atributo: ' + col_name)

        # graph 1
        ax1.title.set_text('Histograma: ' + col_name)
        ax1.hist(base[col_name], bins=1000)
        ax1.set_xlabel('Valores de ' + col_name)
        ax1.set_ylabel('Quantidade de ocorrências')

        # graph 2
        keys = base[col_name].value_counts().keys()
        values = base[col_name].value_counts().values
        ax2.title.set_text('Scatter: ' + col_name)
        ax2.scatter(keys, values)
        ax2.set_xlabel('Valores de ' + col_name)
        ax2.set_ylabel('Quantidade de ocorrências')

        # graph 3
        ax3.title.set_text('Boxplot: ' + col_name)
        ax3.boxplot(base[col_name], meanline=False)

        # statistical text
        s = ''
        for i, j in zip(keys, values):
            s = s + str(i)+': '+str(round(j,2))+'\n'
        props = dict(boxstyle='Sawtooth', facecolor='white', alpha=0.3, pad=.5)
        ax3.text(1.1, 1, s, transform=ax3.transAxes, fontsize=12, verticalalignment='top', bbox=props)
        ax3.set_xlabel('Valores de ' + col_name)
        plt.show()
```

- Ao plotar os atributos qualitativos da mesma maneira que os quantitavos verficou-se que não fez muito sentido plotar de tal modo, por isso usou-se outras formas de plotar os dados qualitativos como os gráficos de barras.

### 4.3.1. Apresente métricas de desempenho compatíveis para a avaliação do problema (liste-as)
- Segue uma lista de métricas que podem ser usadas na avaliação do desempenho das redes neurais para esse problema.
  - **Acurácia**: Ela diz como o modelo classificou corretamente as classes e é dada pela equação:$$ Acurácia = \cfrac{TP + TN}{TP + TN + FP + FN}$$
  - **Precisão**: A precisão é definida pela razão entre a quantidade de exemplos classificados corretamente como positivos e o total de exemplos classificados como positivos, conforme a fórmula abaixo: $$Precisão = \cfrac{TP}{TP + FP}$$
  - **Revocação**: Ao contrário da precisão, a revocação, dá maior ênfase para os erros (falso negativo). Esta métrica é definida pela razão entre a quantidade de exemplos classificados corretamente como positivos e a quantidade de exemplos que são verdadeiramente positivos: $$Revocação = \cfrac{TP}{TP + FN}$$
  - **F1 Score**: A F1 score leva em consideração tanto a precisão quanto a revocação. Ela é definida pela média harmônica entre as duas: $$F_1 = 2*\frac{precisão * recall}{precisão + recall}$$
