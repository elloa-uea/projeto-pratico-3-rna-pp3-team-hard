### Função de ativação sigmoide

$$f(x) = \frac{1}{1 + e^{-\beta \cdot x}}$$

- Implemente uma função em Python que retorne o valor da função sigmóide para uma dada entrada
- Nesta função, considere beta como parâmetro nomeado com valor igual a 1.0
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função sigmoide
- Em um mesmo gráfico, ilustre 3 valores distintos para $\beta$:
  - $\beta = 0.7$ na cor laranja
  - $\beta = 0.5$ na cor verde
  - $\beta = 0.3$ na cor vermelha
  - Use legendas no gráfico

```python
sigmoide = lambda x, beta: (1.0 / (1.0 + np.exp(-beta * x)))
```

```python
plt.plot(entrada, sigmoide(entrada, 1.0))
plt.title('Função Sigmóide')
plt.show()
```

<h3 align='center'>Derivada da função de ativação Sigmoide.</h3>

$$f'(x) = f(x)(1 - f(x))$$

```python
derivada_sigmoide = lambda x, beta: (1.0 / (1.0 + np.exp(-beta * x))) * (1 - (1.0 / (1.0 + np.exp(-beta * x))))
```

```python
plt.plot(entrada, derivada_sigmoide(entrada, 1.0))
plt.title('Derivada Função Sigmóide')
plt.show()
```

```python
for beta, marker in [(0.7, '--'), (0.5, '.'), (0.3, '-')]:
    plt.plot(entrada, sigmoide(entrada, beta), marker, label="beta - %s\n" % beta)

plt.legend()
plt.show()
```

### Função de Ativação ReLU (Unidade Linear Retificada)

$$f(x) = max(0,x)$$

- Implemente uma função em Python que retorne o valor da função ReLU para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função ReLU


```python
relu = lambda x: max(np.max(x), 0.0)
```

```python
plt.plot(entrada, [relu(x) for x in entrada])
plt.title('Função ReLU')
plt.show()
```

<h3 align='center'>Derivada da função de ativação ReLU. </h3>

$$f'(x) = \left \{ \begin{matrix} 0, & \mbox{se }x\mbox{ < 0} \\ 1, & \mbox{se }x \geq \mbox{0}\end{matrix} \right.$$





```python
derivada_relu = lambda x: 1.0 if x >= 0.0 else 0.0
```

```python
plt.plot(entrada, [derivada_relu(x) for x in entrada])
plt.title('Derivada Função ReLU')
plt.show()
```

### Função de ativação linear (purelin)

$$f(x) = x$$

- Implemente uma função em Python que retorne o valor da função linear para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função em questão

```python
purelin = lambda x: x
```

```python
plt.plot(entrada, purelin(entrada))
plt.title('Função Purelin')
plt.show()
```

<h3 align='center'>Derivada da função de ativação Purelin. </h3>

$$f'(x) = 1.x^{1-1}$$


```python
derivada_purelin = lambda x: (1 * (x ** (1-1)))
```

```python
plt.plot(entrada, derivada_purelin(entrada))
plt.title('Derivada Função Purelin')
plt.show()
```

### Função de ativação Tangente Hiperbólica

$$f(x) = tanh(x)$$

- Implemente uma função em Python que retorne o valor da função tangente hiperbólica para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função em questão
- Utilize bibliotecas para auxiliar nesta tarefa
- Considere o intervalo (-1,1)

```python
tanh = lambda x: np.tanh(x)
```

```python
plt.plot(entrada, tanh(entrada))
plt.title('Função Tangente Hiberbólica')
plt.show()
```

<h3 align='center'>Derivada da função de ativação Tangente Hiperbólica. </h3>

$$f'(x) = 1−tanh^2(x)$$


```python
derivada_tanh = lambda x: 1.0 - (np.tanh(x)**2)
```

```python
plt.plot(entrada, derivada_tanh(entrada))
plt.title('Derivada Função Tangente Hiberbólica')
plt.show()
```

### Função de Ativação Gaussiana

$$f(x) = \exp \left( \frac{-(x - \mu)^2}{2\sigma^2} \right)$$

- $\mu$ é o ponto médio (centro) da curva
- $\sigma$ denota o raio de abertura
- Implemente uma função em Python que retorne o valor da função gaussiana para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função em questão


Considere também os seguintes valores ao plotar a função

- Parâmetros nomeados $\mu = 0$ e $\sigma = 1$
- $\mu = 0, \sigma = 0.3$
- $\mu = 0, \sigma = 0.5$
- $\mu = 0, \sigma = 0.7$
- $\mu = 0, \sigma = 0.9$
- $\mu = 0, \sigma = 1.1$

```python
gaussiana = lambda x, u=0, o=1: np.exp((-(x - u )**2) / (2 * (o**2)))
```

```python
for raio, marker in [(0.3, '+'), (0.5, '-'), (0.7, '--'), (0.9, '_'), (1.1, '.')]:
    plt.plot(entrada, gaussiana(entrada, 0, raio), marker, label="raio - %s\n" % raio)
plt.title('Função Gaussiana')
plt.legend()
plt.show()
```

<h3 align='center'>Derivada da função de ativação Gaussiana.</h3>

$$f'(x) = f(x)$$


```python
derivada_gaussiana = lambda x: gaussiana(x)
```

```python
plt.plot(entrada, derivada_gaussiana(entrada), label='Raio - 1\nPonto Médio - 0') 
plt.title('Derivada Função Gaussiana')
plt.legend()
plt.show()
```

### Função de Ativação Leaky ReLU

$$f(x) = \left\{\begin{array}{cc} 0.01\cdot x & x < 0\\
x & x \geq 0
\end{array}
\right.$$

- Implemente uma função em Python que retorne o valor da função Leaky ReLU para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função Leaky ReLU

```python
leaky_relu = lambda x, a=0.01: (a * x) if x < 0.0 else x
```

```python
plt.plot(entrada, [leaky_relu(x) for x in entrada])
plt.title('Função LeaKy ReLU')
plt.show()
```

<h3 align='center'>Derivada da função de ativação Leaky ReLU. </h3>

$$f'(x) = \left\{\begin{array}{cc} 0.01 & x < 0\\
1 & x \geq 0
\end{array}
\right.$$



```python
derivada_leaky = lambda x, a=0.01: 1 if x >= 0 else a
```

```python
plt.plot(entrada, [derivada_leaky(x) for x in entrada])
plt.title('Função LeaKy ReLU')
plt.show()
```
